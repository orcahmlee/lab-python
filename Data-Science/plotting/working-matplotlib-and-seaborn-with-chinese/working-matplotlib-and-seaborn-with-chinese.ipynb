```python
from random import randint
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    '年份': ['2013', '2014', '2015', '2016', '2017', '2019'],
    '價格': [randint(100, 10000) for _ in range(6)]
})
df
```

## Default settings

### matplotlib

```python
plt.plot(
    df['年份'],
    df['價格']
)
plt.xlabel('年份')
plt.ylabel('價格')
plt.savefig('./does-not-show-chinese.png')
```

### Seaborn

```python
sns.lineplot(x="年份", y="價格", data=df)
```

## After change the settings

```python
# find the default font of matplotlib
from matplotlib.font_manager import findfont, FontProperties  


findfont(FontProperties(family=FontProperties().get_family()))
```

```python
# check the default font
import matplotlib.pyplot as plt


plt.rcParams['font.sans-serif']
```

```python
plt.plot(
    df['年份'],
    df['價格']
)
plt.xlabel('年份')
plt.ylabel('價格')
plt.savefig('./show-chinese.png')
```

```python
ax = sns.lineplot(x="年份", y="價格", data=df)
```

```python
sns.set(style='darkgrid', font='SimHei', rc={'figure.figsize':(12, 8)}, font_scale=1.3)
ax = sns.lineplot(x="年份", y="價格", data=df)
ax.figure.savefig('./seaborn-plot.png')
```
